```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Reading our data into a Dataframe

```python
stats_df = pd.read_csv('NBA_Raw_data.csv')
```

```python
stats_df.info()
```

```python
stats_df.describe(include='all')
```

### Removing null values in the Player column

```python
stats_df.drop(stats_df[stats_df['Player'].isnull()==True].index, inplace = True, axis = 0)
```

```python
stats_df.info()
```

### Filling empty stats with the value 0

```python
stats_df.fillna(0, inplace = True)
```

### Handling outliers

```python
plt.figure(figsize=(15,10))
sns.set(font_scale=1)
ax = sns.boxplot(data=stats_df)
```

```python
# Drop all players that didn't play a minute in the season, they will harm our model.
stats_df.drop(stats_df[stats_df['MP'] == 0].index, axis = 0, inplace = True)

# Drop all players that played under 20 games in a season.
stats_df = stats_df[stats_df['G'] >= 20]

# Reset index
stats_df.reset_index(drop = True, inplace = True)

# Let's see our data after removing outliners.
stats_df.describe()
```

```python
# Manually adding the PER value of each player.
# If you want to learn more about the PER value of each player,
# you can read more about it in the following link: https://bleacherreport.com/articles/113144-cracking-the-code-how-to-calculate-hollingers-per-without-all-the-mess

def addPERColumn(df):
    df_with_PER = df.copy()
    df_with_PER['PER'] = 0.0

    for index, row in df_with_PER.iterrows():
        df_with_PER.at[index,'PER'] = round(((row['FG']*85.910) + (row['STL']*53.897) + (row['3P']*51.757) + (row['FT']*46.845) + (row['BLK']*39.190)
                     + (row['ORB']*39.190) + (row['AST']*34.677) + (row['DRB']*14.707) - (row['PF'] * 17.174)
                     - ((row['FTA'] - row['FT'])*20.091) - ((row['FGA']-row['FG'])*39.190) - (row['TOV'] *53.897)) * (1/row['MP']),2)
    return df_with_PER
```

```python
# Adding the PER column to the DataFrame
stats_df = addPERColumn(stats_df)
```

```python
stats_df.info()
```

```python
stats_df.describe()
```

```python
# Adding an 'Efficient' label to determine whether the player was efficient during his season or not.
# We determine it by his PER. If it's >= 15, he is efficient and we fill the cell with '1'. 
# We fill the cell with '0' if his PER is < 15.
stats_df.loc[stats_df['PER'] >= 15, 'Efficient'] = 1 
stats_df.loc[stats_df['PER'] < 15, 'Efficient'] = 0
```

```python
stats_df.info()
```

```python
stats_df
```

### Saving the data

```python
# Saving the DataFrame with the 'Efficient' label into a csv file.
stats_df.to_csv('NBA_Clean_data_Efficient.csv', index=False)

# Saving the DataFrame without the 'Efficient' label.
stats_df[stats_df.columns[~stats_df.columns.isin(['Efficient'])]].to_csv('NBA_clean_data.csv', index=False)
```
