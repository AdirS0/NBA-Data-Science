```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### Reading our data.

```python
final_df = pd.read_csv('NBA_Clean_data.csv')
```

### Let's visualize the correlation data using a heatmap. Focus on the PER column.

```python
plt.figure(figsize=(20,10))
c = final_df.corr()
sns.heatmap(c,cmap='BrBG',annot=True)
```

### Let's take the columns that make the most influence on the PER (corr > 0.65) and present their correlation with a pairplot.

```python
sns.pairplot(final_df[['PER','FG','FGA','2P','FT','FTA','PTS']], kind='kde')
```
